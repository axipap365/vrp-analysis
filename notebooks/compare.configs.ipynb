# Compare costs for different configs

This script visualizes the difference in cost between used configs

```python
import pandas as pd
from extractors import *

df = pd.read_csv('../experiments/demo/out/best_known_solutions.csv')


for problem in get_problems(df):
    columns = []
    for config in get_configs(df):
        columns.append(extract_column(df, config_name = config, problem_name = problem, 
                                      target_column = 'Cost', new_label=config))
    pd.concat(columns, axis = 1).plot.box(title = problem, rot = 45)
```
